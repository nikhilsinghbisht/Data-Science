```python
#Ans2.
DDL stands for Data Definition Language, and it is a subset of SQL (Structured Query Language) used to define the structure and schema 
of a database, as well as to create, modify or delete database objects.
Here are some of the most commonly used DDL commands with examples:

CREATE: The CREATE command is used to create a new database object such as tables, views, indexes, and stored procedures.
Here's an example of creating a table named "customers"

CREATE TABLE customers (
  id INT NOT NULL,
  name VARCHAR(50) NOT NULL,
  email VARCHAR(50),
  PRIMARY KEY (id)
);

2.DROP: The DROP command is used to remove a database object, such as a table or view, from the database.
Here's an example of dropping the "customers" table created in the previous example:

DROP TABLE customers;

3.ALTER: The ALTER command is used to modify the structure of a database object. Here's an example of altering the "customers" table to add a new column named "phone":
sql

ALTER TABLE customers
ADD COLUMN phone VARCHAR(20);

4.TRUNCATE: The TRUNCATE command is used to delete all the data in a table, but the table structure remains intact. 
Here's an example of truncating the "customers" table:

TRUNCATE TABLE customers;
```

```python
#Ans3.
DML stands for Data Manipulation Language, which is a subset of SQL used to manipulate data in a database. 
DML commands are used to insert, update, and delete data in a table.

Here are some of the most commonly used DML commands with examples:
INSERT: The INSERT command is used to add new data to a table. 
Here's an example of inserting a new record into a "customers" table:
INSERT INTO customers (id, name, email)
VALUES (1, 'John Doe', 'johndoe@example.com');

2.UPDATE: The UPDATE command is used to modify existing data in a table. 
Here's an example of updating the email address of a customer with an id of 1:
UPDATE customers
SET email = 'john.doe@example.com'
WHERE id = 1;

3.
DELETE: The DELETE command is used to remove data from a table.
Here's an example of deleting a customer with an id of 1:
DELETE FROM customers
WHERE id = 1;

```

```python
#Ans4.
DQL stands for Data Query Language, and it is a subset of SQL (Structured Query Language) used to query and retrieve data from a database. 
The most commonly used DQL command is the SELECT statement, which allows you to retrieve data from one or more tables in a database.
Here's an example of how to use the SELECT statement to retrieve data from a table:

Let's say you have a table named "employees" with columns "id", "name", "department", "position", and "salary".
To retrieve all the data from this table, you would use the following SELECT statement:
SELECT * FROM employees;

The asterisk (*) symbol is a wildcard character that indicates all columns should be selected. 
If you want to select specific columns, you can list them after the SELECT statement, like this:
SELECT id, name, salary FROM employees;




  
```

```python
#Ans5.
Primary key and foreign key are important concepts in relational databases.

A primary key is a column or a set of columns in a table that uniquely identifies each row in the table. 
It must have a unique value for each row and cannot be null. A primary key is used to enforce data integrity and ensure that
there are no duplicate rows in the table. It is often used as a reference for other tables in a database.

Here is an example of creating a table with a primary key:
  CREATE TABLE students (
  id INT PRIMARY KEY,
  name VARCHAR(50),
  age INT
);
In this example, the "id" column is the primary key. It will uniquely identify each row in the "students" table.

A foreign key is a column or a set of columns in a table that refers to the primary key of another table. 
It establishes a link between two tables and ensures referential integrity between them. 
The values in the foreign key column must match the values in the primary key column of the other table. 
This constraint ensures that the data in the foreign key column is valid and refers to an existing row in the other table.

Here is an example of creating a table with a foreign key:
CREATE TABLE courses (
  id INT PRIMARY KEY,
  name VARCHAR(50),
  teacher VARCHAR(50)
);

CREATE TABLE enrollments (
  student_id INT,
  course_id INT,
  FOREIGN KEY (student_id) REFERENCES students(id),
  FOREIGN KEY (course_id) REFERENCES courses(id)
);
In this example, the "enrollments" table has two foreign keys: "student_id" and "course_id". 
They both reference the primary key of their respective tables, "students" and "courses". 
This ensures that the values in the foreign key columns refer to existing rows in the other tables.

In summary, a primary key uniquely identifies each row in a table, while a foreign key establishes a link between two tables and
ensures referential integrity. 
These constraints are important for maintaining data integrity and preventing errors in a relational database.

```

```python
#Ans6.

import mysql.connector

# Connect to the database
mydb = mysql.connector.connect(
  host="localhost",
  user="yourusername",
  password="yourpassword",
  database="yourdatabase"
)

# Create a cursor
mycursor = mydb.cursor()

# Execute a query
mycursor.execute("SELECT * FROM yourtable")

# Fetch the results
results = mycursor.fetchall()

# Print the results
for result in results:
  print(result)

'''The mysql.connector module is used to connect to the MySQL database. In the code above, we specify the host, user, password, and 
database that we want to connect to.

Once we've established the connection, we create a cursor object using the cursor() method. The cursor is a control structure that
lets you execute SQL queries against the database and fetch the results.

The execute() method of the cursor object is used to execute a SQL query. In the example above, we're executing a simple SELECT query
to fetch all the rows from a table. You can execute any SQL query using this method.

Finally, we fetch the results using the fetchall() method, which returns a list of tuples containing the data fetched from the query.'''
```

```python
#Ans 7

In an SQL query, the clauses are executed in a specific order:

FROM: The FROM clause specifies the table or tables from which to retrieve data.

JOIN: The JOIN clause is used to combine rows from two or more tables, based on a related column between them.

WHERE: The WHERE clause is used to filter the results of the query based on a specific condition. It is executed after the tables are joined.

GROUP BY: The GROUP BY clause is used to group the results by one or more columns. It is executed after the WHERE clause.

HAVING: The HAVING clause is used to filter the results of a GROUP BY clause based on a specific condition. It is executed after the
GROUP BY clause.

SELECT: The SELECT clause is used to select the columns to be retrieved from the tables specified in the FROM clause. 
It is executed after all previous clauses.

ORDER BY: The ORDER BY clause is used to sort the results by one or more columns. It is executed after all previous clauses.

LIMIT/OFFSET: The LIMIT and OFFSET clauses are used to limit the number of rows returned by the query and to start retrieving results
from a specific row. It is executed after all previous clauses.
```
