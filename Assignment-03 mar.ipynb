```python
#Ans 1

Five plots in Seaborn along with their usage is as following :
    
Box Plot : A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers.
It is used to identify the spread and skewness of the data.

Heatmap : A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to
identify patterns in the data, such as correlations or clusters.

Pair Plot : A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used
to identify correlations between variables and to explore the relationships between them.

Violin Plot : A violin plot is a graphical representation of data that combines elements of a box plot and a kernel density plot. It is used to
visualize the distribution of a dataset and show its density.

Joint Plot : A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.
```

```python
#Ans 2

import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
sns.lineplot(data=fmri,x='timepoint',y='signal',style='region',hue='event')
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

```python
#Ans 3

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Showing the boxplot for pclass vs age

sns.boxplot(data=titanic,x='pclass',y='age')
```

```python
# Showing the boxplot for pclass vs fare

sns.boxplot(data=titanic,x='pclass',y='fare')
```

```python
#Ans 4

diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=diamonds, x='price',hue='cut',kde=True)
plt.title('Distribution of Price for various diamond cut')
plt.show()
```

```python
#ans 6

flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

```python
sns.heatmap(flights_matrix, 
            cmap="YlGnBu", 
            annot=True, 
            fmt="d", 
            linewidths=.5,
            linecolor="black")
plt.title("Passengers per Month")
plt.show()
```
